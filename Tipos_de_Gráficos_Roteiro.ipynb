#Crie uma função para realizar o gráfico de linhas passando como parâmetro o estilo 

# def GraficoLinha(style)

```python
import matplotlib.pyplot as plt

def GraficoLinha(style='ggplot'):
    # Estilo de cor
    plt.style.use(style)

    # Aumentar o tamanho da fonte
    plt.rcParams.update({'font.size': 16})

    # Dados fictícios para exemplo
    df = {
        'Meses': ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio'],
        'Produto 1': [100, 150, 120, 200, 180],
        'Produto 2': [90, 120, 140, 110, 160],
        'Produto 3': [80, 110, 130, 150, 120]
    }

    # Criar um gráfico de linha
    plt.figure(figsize=(16, 12))

    # Descompactar lista
    produto1, = plt.plot(df['Meses'], df['Produto 1'], label='Produto 1', marker='o')
    produto2, = plt.plot(df['Meses'], df['Produto 2'], label='Produto 2', marker='o')
    produto3, = plt.plot(df['Meses'], df['Produto 3'], label='Produto 3', marker='o')

    plt.legend([produto1, produto2, produto3], ['Produto 1', 'Produto 2', 'Produto 3'])
    plt.title('Vendas de Produtos (2023)')
    plt.xlabel('Mês')
    plt.ylabel('Vendas')

    plt.savefig('GraficoLinha.svg')
    # Mostrar o gráfico
    plt.show()

GraficoLinha('Solarize_Light2')
```

```python
import matplotlib.pyplot as plt

def scatter_plot(df):
    plt.scatter(df['Idade'], df['Preço'])
    
    plt.title('Preço do Carro x Idade')
    plt.xlabel('Idade do Carro (Anos)')
    plt.ylabel('Preço (R$)')

    # Mostrar o gráfico
    plt.show()
```

# Explique como esse gráfico:

# Por que é interessante fazer um gráfico de dispersão nesse caso?
R: Fazer um gráfico de dispersão nesse caso é interessante porque permite visualizar a relação entre duas variáveis quantitativas, no caso, a idade do carro e o preço. O gráfico de dispersão é uma representação gráfica simples, onde cada ponto no gráfico representa uma observação individual com os valores correspondentes de ambas as variáveis.

# Qual é a diferença entre o gráfico de linha na interpretação?
R:  O gráfico de dispersão é usado para analisar a relação entre duas variáveis contínuas, enquanto o gráfico de linha é usado para visualizar a evolução ou mudança de uma variável ao longo do tempo ou em uma sequência ordenada. Cada tipo de gráfico tem sua própria aplicação e interpretação específica, dependendo do contexto e das perguntas que se deseja responder.



# Historiograma

## Modifique a distribuição aleatória para distribuição normal - Pesquise a função numpy

```python
import numpy as np
import matplotlib.pyplot as plt

# Vamos gerar um conjunto de dados aleatórios com distribuição normal para as idades
np.random.seed(0)
idades = np.random.normal(loc=50, scale=10, size=1000)  # distribuição normal

# Criar o histograma
plt.hist(idades, bins=20, edgecolor='black')

plt.title('Distribuição das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

# Interprete a distribuição distribuição e compare a diferença com a distribuição normal

R: A distribuição das idades segue uma distribuição normal, onde a maioria das idades está próxima da média e diminui à medida que nos afastamos dela. Isso pode ser útil para analisar tendências, criar modelos estatísticos ou entender o comportamento das idades em um contexto específico.

# Gráfico de Caixa (Box Plot)

```python
# Dados
distribuicao = np.random.normal(40, 20, 120)

# Criar o gráfico de caixa
plt.boxplot(distribuicao)

plt.title('Distribuição Personalizada')
plt.xlabel('Dados')
plt.ylabel('Valores')

# Mostrar o gráfico
plt.show()
```

# Explique o Gráfico acima -- Pesquise
R: Esse gráfico de caixa é útil para visualizar a distribuição dos dados e identificar valores extremos (outliers). Ele fornece informações sobre a tendência central dos dados (mediana), a dispersão (IQR) e a presença de valores discrepantes. É especialmente útil para comparar várias distribuições lado a lado ou para identificar valores atípicos em um conjunto de dados.

```python
df = pd.read_csv('Pontuacao.csv')
df
```

```python
# Criar o gráfico de violino
equipes = df.columns.tolist()
plt.figure(figsize=(8, 6))
plt.violinplot(df)

# Personalizar o gráfico
plt.title('Distribuição da Pontuação por Equipe')
plt.xlabel('Equipes')
plt.ylabel('Pontuação')

plt.xticks(np.arange(1, len(equipes)+1), equipes)

# Mostrar o gráfico
plt.show()
```

# Explique o gráfico acima, qual é a diferença da Equipe C da Equipe D.
R: Com base no gráfico de violino, podemos concluir que a Equipe C tem uma distribuição mais concentrada e com pontuações mais altas em relação à Equipe D. A Equipe D, por sua vez, apresenta uma distribuição mais dispersa, com uma maior variação nas pontuações e valores mais baixos predominantes.
