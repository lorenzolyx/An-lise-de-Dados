# Qual a linha de produtos tem a melhor avaliação?

```python
Agrupamento = df.groupby('Linha de produto')['Avaliação'].mean().reset_index(drop=False)
```

```python
Agrupamento
```

```python
Agrupando_avaliação = Agrupamento.round(1)
```

```python
Agrupando_avaliação
```

```python
Agrupando_avaliação = Agrupando_avaliação.sort_values(by = 'Avaliação', ascending = False).
```

```python
Figura = sns.color_pallete('gnuplot')
Figura.reverse()
```

```python
ax = sns.barbloty = 'Linha de produto', x = 'Avaliação', data = Agrupando_avaliação, palette = Figura)
```
